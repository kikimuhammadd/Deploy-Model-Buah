```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.python.ops.numpy_ops import np_config
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/Gizi_Buah.csv', on_bad_lines='warn')
df
```

```python
df['Buah'] = df['Buah'].str.strip()
```

```python
# Mengambil nama buah
fruits_names = df['Buah'].values

# Mengambil kolom-kolom yang ingin dijadikan features
data_fruits = df[['Kalori', 'Lemak(g)', 'Karbohidrat(g)', 'Protein(g)']].values
```

```python
# Normalisasi data
data_normalisasi = (data_fruits - np.min(data_fruits, axis=0)) / (np.max(data_fruits, axis=0) - np.min(data_fruits, axis=0))
# Membagi data menjadi data pelatihan, validasi, dan uji
train_data, test_data, train_labels, test_labels = train_test_split(data_normalisasi, data_normalisasi[:, 0], test_size=0.2, random_state=42)
train_data, validation_data, train_labels, validation_labels = train_test_split(train_data, train_labels, test_size=0.2, random_state=42)
```

```python
# Membangun model TensorFlow
model_fruits= tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
model_fruits.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Melatih model
model_fruits.fit(train_data, train_labels, validation_data=(validation_data, validation_labels), epochs=300, verbose=1)
```

```python
loss = model_fruits.evaluate(test_data, test_labels)
```

```python
from tensorflow.keras.models import save_model
model_path = '/content/drive/MyDrive/Buah_dan_Gizi_Final.h5'
save_model(model_fruits, model_path)
```

```python
tflite_model = tf.keras.models.load_model("/content/drive/MyDrive/Buah_dan_Gizi_Final.h5")
converter = tf.lite.TFLiteConverter.from_keras_model(tflite_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_save = converter.convert()
open("/content/drive/MyDrive/model_buah_final.tflite", "wb").write(tflite_save)
```

```python
import matplotlib.pyplot as plt
from google.colab import files
from keras.preprocessing import image
from tensorflow.python.ops.numpy_ops import np_config
# Mengunggah foto ke file lokal
uploaded = files.upload()


# Mengambil path foto yang telah diunggah
image_path = list(uploaded.keys())[0]

#Membuat variabel untuk model training
modell = tf.keras.models.load_model('/content/drive/MyDrive/final.h5')
model = modell

# Daftar nama class pada setiap buah
class_names = ["Apel","Anggur","Buah Naga","Ceri", "Durian", "Jambu Biji","Jeruk", "Kiwi", "Lemon", "Mangga", "Nanas", "Pir","Pisang", "Semangka", "Stroberi" ]

# Melakukan persiapan untuk memprediksi foto yang telah diunggah
img = tf.keras.utils.load_img(image_path, target_size=(150, 150))
x = tf.keras.utils.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = x / 255.0

# Convert foto ke dalam tensor
x = tf.convert_to_tensor(x, dtype=tf.float32)

# Prediksi pada foto yang telah diunggah
predictions = model(x)
predicted_class_index = tf.argmax(predictions[0]).numpy()
predicted_class = class_names[predicted_class_index]

# Mempersiapkan info gizi pada buah
predicted_class = class_names[predicted_class_index]
nutrient_info = df.loc[df['Buah'] == predicted_class, ['Kalori', 'Lemak(g)', 'Karbohidrat(g)', 'Protein(g)','Ukuran']]

# Menampilkan hasil prediksi
print('Predicted class:', predicted_class)

# Menampilkan informasi nutrisi tanpa indeks
print("Nutrient Information:")
print(nutrient_info.to_string(index=False))

# Menampilkan gambar
img = tf.keras.utils.load_img(image_path)
plt.imshow(img)
plt.axis('off')
plt.show()
```
